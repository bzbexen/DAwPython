## Excel
Pandas can read and write excel files, keep in mind, this only imports data. Not formulas or images, having images or macros may cause this read_excel method to crash. 

### Excel Input

```python
df1=pd.read_excel("Excel_Sample.xlsx")  # default sheet1 getirir
df1
```

```python
df2=pd.read_excel("Excel_Sample.xlsx")
df2
```

```python
df3=pd.read_excel("Excel_Sample.xlsx", sheet_name=None)
# sheet_name=None : butun sheetleri acar
df3
```

```python
type(df3)
```

```python
pd.ExcelFile('Excel_Sample.xlsx').sheet_names
#  sheet nameleri verir 
```

### Excel Output

```python
df1.to_excel("Excel_Sample.xlsx", sheet_name="sheet1", index=False)
```

```python
df2.to_excel("Excel_Sample.xlsx", sheet_name="sheet2", index=False) 
#sheet2'yi' tum dosyaya kaydetti (eskiden dosya icinde baska sheetler varsqa hepsi gider)!!!
```

```python
with pd.ExcelWriter('combined_df.xlsx') as writer:  #ayri ayri sheet olusturur
    df1.to_excel(writer, sheet_name='sample1', index=False) # indexfalse  
    df2.to_excel(writer, sheet_name='sample2', index=True) # indexi ekler 
```

# SQL Connections


```python
# pip install sqlalchemy
```

```python
from sqlalchemy import create_engine # sql ile baglantiyi kuracak tool create_engine
```

```python
temp_db = create_engine("sqlite:///:memory:") #baglanti kurmak icin baglanti kuruyoruz 
#sqlitein python kullanimi cok rahat 
# bilg rem icinde gecici database olusturur, arasak bulamayiz 
# memory yerine baska bir sey olabilir, ama genel kullanim "memory"
```

```python
temp_db.table_names()
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')
# istedigimiz sayfadan dosyayi, tables dataframe icine attik 
```

```python
tables
```

```python
tables[6]
```

```python
tables[6].to_sql(name="populations", con=temp_db)
# populations tablo ismiyle temp_db 'ye atcak. 
```

```python
temp_db.table_names() # temp_db icinde yer alan table isimlerini getirir 
```

```python
# sqlde populations table'i temp_db  icinden  getir 
df = pd.read_sql(sql="populations", con= temp_db)
```

```python

pd.read_sql_query(sql="SELECT Country, Population FROM populations", con=temp_db)
```
